```python
# Importing libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
# To get 1st five rows of data
df=pd.read_csv('data.csv')
df.head()
```

```python
# To get last five rows of data
df.tail()
```

```python
# To get no.of columns and rows
df.shape
```

```python
# To get some information about the datatset
df.info()
```

```python
# Describe used for to get statistical analysis
df.describe()
```

```python
# To get columns names
df.columns
```

```python
# Checking if there any duplicated present or not
df.duplicated().sum()
```

```python
# Checking if there any null values present or not
df.isnull().sum()
```

```python
# Assigning df to an new file cleaned_df for performing data wrangling
cleaned_df=df
# Use to_csv function to write and save as a file
cleaned_df.to_csv('cleaned_data.csv',index=False)
```

```python
# Checking if there any null values present or not
cleaned_df.isnull().sum()
```

```python
# Setting a random seed ensures reproducibility
np.random.seed(42)
# Generate random heights between 150 and 180 cm for all employees
cleaned_df['Height'] = np.random.randint(150, 181, df.shape[0])
cleaned_df.head()
```

```python
# Checking median of the salary to replace missing values
cleaned_df['Salary'].median()
```

```python
# Replace missing salaries with the median salary for consistency
median_salary = cleaned_df['Salary'].median()
cleaned_df['Salary'] = cleaned_df['Salary'].fillna(median_salary)
cleaned_df.head()
```

```python
# Checking mode of college to replace the missing values
cleaned_df['College'].mode()
```

```python
# Replace missing college names with the most common value (mode)
mode_college = cleaned_df['College'].mode()[0]
cleaned_df['College'] = cleaned_df['College'].fillna(mode_college)
cleaned_df['College']
```

```python
# Checking after performing some data wrangling methods
cleaned_df.head()
```

```python
# Checking if there any null values
cleaned_df.isnull().sum()
```

```python
# Finding percentage_split to perform piechart
percentage_split=cleaned_df['Team'].value_counts() / len(cleaned_df) * 100
percentage_split
```

```python
# Piechart represent the percentage_split
plt.figure(figsize=(9, 7))
plt.pie(percentage_split,labels=percentage_split.index)
plt.title('Distribution of Employees Across Teams')
plt.tight_layout()
plt.show()
```

```python
# Finding the position and count
position_group = cleaned_df.groupby('Position').size().reset_index(name='Count')
position_group
```

```python
# Barh repersent the position and count of employees
plt.figure(figsize=(8, 6))
plt.barh(position_group['Position'], position_group['Count'], color='skyblue',edgecolor='black')
plt.xlabel('Number of Employees')
plt.ylabel('Position')
plt.title('Number of Employees in Each Position')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Checking age of employee
age_count=cleaned_df['Age'].value_counts()
age_count
```

```python
# Return index of occurrence of maximum value
age_count.idxmax()
```

```python
# Plot histogram with the finding things
plt.figure(figsize=(8, 6))
plt.hist(cleaned_df['Age'], bins=20, color='brown', edgecolor='black')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Predominant Age Group')
plt.grid(alpha=0.1)
plt.tight_layout()
plt.show()
```

```python
# Checking salary of employee with value.counts()
salary=cleaned_df['Salary'].value_counts()
salary
```

```python
# Finding which team and position has high salary from the employee
salary=cleaned_df.groupby(['Team','Position'])['Salary'].sum()
highest_salary=salary.idxmax()
print(f'highest_salary',highest_salary)
print(f'Team',highest_salary[0])
print(f'Position',highest_salary[1])
```

```python
# Use pivot functions to create a table for team , position and salary
salary_pivot=salary.reset_index().pivot_table(index='Team',columns='Position',values='Salary')

# With pivot ploting bar chart to perform the Total Salary Expenditure by Team and Position
ax = salary_pivot.plot(
    kind='bar',
    figsize=(14, 8),
    stacked=False,
    alpha=0.8,
    color=sns.color_palette('Paired', n_colors=salary_pivot.shape[1]))
plt.title('Total Salary Expenditure by Team and Position', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Total Salary', fontsize=12)
plt.xticks( fontsize=10)
plt.legend(title='Position', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation between employee age and salary
correlation = cleaned_df['Age'].corr(cleaned_df['Salary'])

# Display and interpret the correlation
print(f"Correlation between Age and Salary: {correlation:.2f}")
if correlation > 0.5:
    print("There is a strong positive correlation between Age and Salary.")
elif correlation < -0.5:
    print("There is a strong negative correlation between Age and Salary.")
else:
    print("The correlation between Age and Salary is weak.")
```

```python
# Plotting scatterplot to perform age vs salary
plt.figure(figsize=(10, 8))
sns.color_palette("crest", as_cmap=True)
sns.scatterplot(x='Age',y='Salary',data=cleaned_df,hue='Age',palette='crest')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(title='Salary',loc='best')
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()
```

**Salary Spending:**

**Age vs. Salary:**
